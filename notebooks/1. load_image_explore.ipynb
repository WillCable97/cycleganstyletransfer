```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from cycleganstyletransfer.config import DATA_DIR
```

```python
data_dir = DATA_DIR / "raw"
monet_dir = data_dir / "Monet"
photo_dir = data_dir / "Images"


# Get list of all files in both directories
monet_files = list(monet_dir.glob('*.jpg'))
photo_files = list(photo_dir.glob('*.jpg'))
```

## Viewing some random selections

```python
# Randomly select one image from each directory
monet_img_path = np.random.choice(monet_files)
photo_img_path = np.random.choice(photo_files)

# Read the images using PIL instead of cv2
monet_img = np.array(Image.open(monet_img_path))
photo_img = np.array(Image.open(photo_img_path))

# Display the images side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
ax1.imshow(monet_img)
ax1.set_title('Monet Image')
ax1.axis('off')
ax2.imshow(photo_img)
ax2.set_title('Photo Image') 
ax2.axis('off')
plt.show()
```
